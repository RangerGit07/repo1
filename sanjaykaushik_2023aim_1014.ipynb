```python
class TreeNode:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.left = None
        self.right = None

class TwoDimBinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, x, y):
        self.root = self._insert_recursive(self.root, x, y)

    def _insert_recursive(self, node, x, y):
        if node is None:
            return TreeNode(x, y)

        if x < node.x:
            node.left = self._insert_recursive(node.left, x, y)
        else:
            node.right = self._insert_recursive(node.right, x, y)

        return node




```
